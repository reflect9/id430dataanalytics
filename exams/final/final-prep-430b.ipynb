## W9-10. Probability, Bayes Theorem, and Causal Analysis
- Basic Concepts of Probability and Compound Probability
    - With given text description / data, you will be asked to tell what is the probability of specific event
        - E.g. What is the probability of getting 4 and 5 if you throw a dice twice?  -> The answer is 1/36  
        - E.g. What is the probability of getting 9 for sum of the eyes if you throw a dice twice? -> Hint. Binomial distribution
        - E.g. If you randomly choose a person in the dataset, what is the probability of choosing 10-19yrs old male?
        
- Conditional Probability and Contingency Table
    - What is the probability of (A to be true if B is false), according to the given contingency table?
        - The contingency table might have missing cells that you need to fill
- Bayes Theorem
    - Based on the given text description of situation, fill in the contingency table and answer the question.
        - E.g. By analyzing people at front of the store (either guests or passer-by) assume that you figured out the following facts: 
            - __Fact.__ Based on statistics, 1 out of 4 people in the town wore coats today
            - __Fact.__ We see, among the guests (who entered the store), 10% wore coats
            - __Fact.__ Half of the people at the front of the store are guests
            - __Question.__ If you see a person wearing a coat, what is the probability of he/she would enter the store?  -> __The answer is 20%__
        - E.g. By combining a customer satisfaction report and log data, we figured out the following facts:
            - __Fact.__ Around 20% of the total customers were not satisfied with our service
            - __Fact.__ Log data tells that 30% of overall customers stay at least five minutes per visit
            - __Fact.__ Log data also tells that only 10% of unsatisfied customers also stay at least five minutes per visit
            - __Question.__ If you see a person who stay five minutes or less, what is the probability of he/she would be unsatisfied?  -> __The answer is 52/70 => 71.4%__. By knowing the duration of stay, we can update the probability of the customer's satisfaction from 80% to 71.4%
